__Prepare DataLoader__ <br>
DataLoader turns out dataset into a Python iterable. More specifically, it turns our data into batches (mini-batches)

```python
from torch.utils.data import DataLoader

# for taking different batch_size change this hyper parameter
BATCH_SIZE = 32

# train_dataloader
train_dataloader = DataLoader(dataset=train_data, batch_size=BATCH_SIZE, shuffle=True)

# test_dataloader
test_dataloader = DataLoader(dataset=test_data, batch_size=BATCH_SIZE, shuffle=False)
```

## Building a model
We will start by creating a baseline model. <br>
A baseline model is a simple model you will try and improve upon with subsequent models/experiments

```python
# checking what would a flatten layer would do
flatten_model = nn.Flatten()

# getting a single image from the batch
x = train_features_batch[0]

# flatten the image
output = flatten_model(x)

print(f"Before Flattening Img shape: {x.shape} -> [color_channels, height, width]")
print(f"After Flattening Img shape: {output.shape} -> [color_channels, height*width]")
```

```python
class FashionMNISTModelV0(nn.Module):
    def __init__(self, input_shape, hidden_units, output_shape):
        super().__init__()
        self.layer_stack = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=input_shape, out_features=hidden_units),
            nn.Linear(in_features=hidden_units, out_features=output_shape)
        )
        
    def forward(self, x):
        return self.layer_stack(x)
```

```python
torch.manual_seed(42)

# if you want to change no of neurons change this hyper parameter
HIDDEN_UNITS = 16

model_0 = FashionMNISTModelV0(
    input_shape = x.shape[1] * x.shape[2], # 28 * 28 = 784
    hidden_units = HIDDEN_UNITS,
    output_shape = len(class_names)
)

model_0
```

```python
# lets see the output of the model_0 on dummy img
dummy_x = torch.rand([1, 1, 28, 28])
model_0(dummy_x)
```

__setting up loss func and optimizer__

```python
# loss func
loss_fn = nn.CrossEntropyLoss()

# optimizer
optimizer = torch.optim.Adam(params=model_0.parameters())
```

```python
# for evaluating the model we use accuracy as metric
def accuracy_fn(y_true, y_preds):
    correct = torch.eq(y_true, y_preds).sum().item()
    return (correct/len(y_true)) * 100
```

__creating a func to time the model__

```python
from timeit import default_timer as timer
def print_train_time(start:float, end:float, device:torch.device = None):
    total_time = end - start
    print(f"Training Time on {device}: {total_time: .3f} seconds")
    return total_time

start_time = timer()
end_time = timer()

print_train_time(start=start_time, end=end_time, device="cpu")
```

__creating a training loop__

```python
# import tqdm for progress bar
from tqdm.auto import tqdm

# set the seed and start the timer
device = "cpu"
torch.manual_seed(42)
start_cpu_train_time = timer()

# change number of epochs by changing this hyper parameter
EPOCHS = 3

for epoch in tqdm(range(EPOCHS)):
    train_loss = 0
    
    for batch, (X, y) in enumerate(train_dataloader):
        model_0.train()
        
        # forward pass
        y_pred = model_0(X)
        
        # calc loss
        loss = loss_fn(y_pred, y)
        train_loss += loss # accumulating the loss per batch
        
        # optimizer zero grad
        optimizer.zero_grad()
        
        # loss backward
        loss.backward()
        
        # optimizer step
        optimizer.step()
        
    train_loss /= len(train_dataloader)
    
    # Testing loop
    test_loss, test_acc = 0, 0
    
    model_0.eval()
    with torch.inference_mode():
        for X_test, y_test in test_dataloader:
            # forward pass
            test_pred = model_0(X_test)
            
            # calc loss, acc
            test_loss += loss_fn(test_pred, y_test)
            
            test_acc += accuracy_fn(y_true=y_test, y_preds=test_pred.argmax(dim=1))
            
        test_loss /= len(test_dataloader)
        test_acc /= len(test_dataloader)
        
    print(f"Epoch: {epoch}\tTrain_Loss:{train_loss}\tTest_Loss:{test_loss}\tTest_Acc:{test_acc}")
    
# stop the timer
end_cpu_train_time = timer()
print_train_time(start=start_cpu_train_time, end=end_cpu_train_time, device="cpu")
```

```python
# function for evaluating the model
torch.manual_seed(42)

def evaluate_model(model:torch.nn.Module, 
                   data_loader: torch.utils.data.DataLoader, 
                   loss_fn: torch.nn.Module, 
                   accracy_fn):
    loss, acc = 0, 0
    model.eval()
    with torch.inference_mode():
        for X, y in data_loader:
            X = X.to(device)
            y = y.to(device)
            # forward pass
            y_pred = model(X)

            # calc loss, acc
            loss += loss_fn(y_pred, y)

            acc += accuracy_fn(y_true=y, y_preds=y_pred.argmax(dim=1))

        loss /= len(data_loader)
        acc /= len(data_loader)
        
    return {'model_name': model.__class__.__name__, # only works when model was created with a class
            'model_loss': loss.item(),
            'model_acc': acc}
```

```python
# checking model_0 results
model_0_results = evaluate_model(model_0, test_dataloader, loss_fn, accuracy_fn)
model_0_results
```

__Creating  model_1__ : with non linearities and on gpu

```python
# device agnostic code
device = "mps" if torch.backends.mps.is_available() else "cpu"
device
```

```python
class FashionMNISTV1(nn.Module):
    def __init__(self, input_features, hidden_units, output_features):
        super().__init__()
        self.layer_stack = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=input_features, out_features=hidden_units),
            nn.ReLU(),
            nn.Linear(in_features=hidden_units, out_features=output_features)
        )
        
    def forward(self, x):
        return self.layer_stack(x)
```

```python
# creating instance

INPUT_FEATURES = x.shape[1] * x.shape[2]
HIDDEN_UNITS = 16
OUTPUT_FEATURES = len(class_names)


model_1 = FashionMNISTV1(INPUT_FEATURES, HIDDEN_UNITS, OUTPUT_FEATURES).to(device)
model_1
```

```python
# loss func
loss_fn = nn.CrossEntropyLoss()

# optimizer
optimizer = torch.optim.Adam(params=model_1.parameters())
```

__functionizing training & testing loop__

```python
def train_step(model: torch.nn.Module,
               data_loader: torch.utils.data.DataLoader,
               loss_fn: torch.nn.Module,
               optimizer: torch.optim.Optimizer,
               accuracy_fn,
               device: torch.device = device):
    
    train_loss, train_acc = 0, 0
    
    model.train()
    
    for X_train, y_train in data_loader:
        # sending batches to the gpu
        X_train = X_train.to(device)
        y_train = y_train.to(device)
        
        # forward pass
        y_preds = model(X_train)
        
        # calc loss, acc
        loss = loss_fn(y_preds, y_train)
        train_loss += loss.item()
        train_acc += accuracy_fn(y_train, y_preds.argmax(dim=1))
        
        # optimizer zero grad
        optimizer.zero_grad()
        
        # loss backwards
        loss.backward()
        
        # optimizer step
        optimizer.step()
        
    train_loss /= len(data_loader)
    train_acc /= len(data_loader)
    
    return train_loss, train_acc
```

```python
def test_step(model: torch.nn.Module,
              data_loader: torch.utils.data.DataLoader,
              loss_fn: nn.Module,
              accuracy_fn,
              device: torch.device = device):
    test_loss, test_acc = 0, 0
    model.eval()
    with torch.inference_mode():
        for X_test, y_test in data_loader:
            X_test = X_test.to(device)
            y_test = y_test.to(device)
            
            # forward pass
            test_preds = model(X_test)
            
            # calc loss, acc
            test_loss += loss_fn(test_preds, y_test).item()
            test_acc += accuracy_fn(y_test, test_preds.argmax(dim=1))
            
        test_loss /= len(data_loader)
        test_acc /= len(data_loader)
    
    return test_loss, test_acc
```

```python
# training 
torch.manual_seed(42)
torch.mps.manual_seed(42)
start_gpu_train_time = timer()

EPOCHS = 3

for epoch in tqdm(range(EPOCHS)):
    train_loss, train_acc = train_step(model_1, train_dataloader, loss_fn, optimizer, accuracy_fn)
    test_loss, test_acc = test_step(model_1, test_dataloader, loss_fn, accuracy_fn)
    
    print(f"Epoch:{epoch}\tTrain_Loss:{train_loss:.4f}\tTrain_Acc:{train_acc:.4f}\tTest_Loss:{test_loss:.4f}\tTest_Acc:{test_acc:.4f}")
    
# stop the timer
end_gpu_train_time = timer()
print_train_time(start=start_gpu_train_time, end=end_gpu_train_time, device="cpu")
```

```python
# saving results of model_1
model_1_results = evaluate_model(model_1, test_dataloader, loss_fn, accuracy_fn)
model_1_results
```

## Building A Convolutional Neural Network
To see what's happening inside a CNN, see the below website<br>
__https://poloclub.github.io/cnn-explainer/__

```python
class FashionMNISTCNN(nn.Module):
    def __init__(self, input_shape:int, hidden_units:int, output_shape:int):
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape,
                      out_channels=hidden_units,
                      kernel_size=3,
                      stride=1,
                      padding=1
            ),
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units,
                      out_channels=hidden_units,
                      kernel_size=3,
                      stride=1,
                      padding=1
                
            ),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )
        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(in_channels=hidden_units,
                      out_channels=hidden_units,
                      kernel_size=3,
                      stride=1,
                      padding=1
            ),
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units,
                      out_channels=hidden_units,
                      kernel_size=3,
                      stride=1,
                      padding=1
            ),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )
        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features= hidden_units*7*7, out_features=output_shape)
        )
        
    def forward(self, x):
        x = self.conv_block_1(x)
        x = self.conv_block_2(x)
        return self.fc(x)
```

```python
# create instance of the model

input_shape = 1 # because our image is in grayscale
HIDDEN_UNITS = 10
output_shape = len(class_names)

cnn = FashionMNISTCNN(input_shape=input_shape, hidden_units=HIDDEN_UNITS, output_shape=output_shape)
cnn
```

__Making Predictions__

```python
def make_predictions(model: torch.nn.Module, data: list, device: torch.device = device):
    pred_probs = []
    model = model.to(device)
    model.eval()
    with torch.inference_mode():
        for sample in data:
            # prepare the sample (add a batch dimension and pass to target device)
            sample = torch.unsqueeze(sample, dim=0).to(device)
            
            # forward pass
            pred_logit = model(sample)
            
            # get prediction probability
            pred_prob = torch.softmax(pred_logit.squeeze(), dim=0)
            
            # get pred_prob off the gpu for further calculations
            pred_probs.append(pred_prob.cpu())
            
    # stack the pred_probs to turn list into a tensor
    return torch.stack(pred_probs)
```

```python
import random
# random.seed(42)

test_samples = []
test_labels = []

for sample, label in random.sample(list(test_data), k=9):
    test_samples.append(sample)
    test_labels.append(label)
    
# getting predictions
pred_probs = make_predictions(model=cnn, data=test_samples)
pred_probs[0]
```

```python
# to get the label we have to use argmax()
pred_labels = pred_probs.argmax(dim=0)
pred_labels
```

```python
# now lets visualize
# plotting random images
torch.manual_seed(42)
fig = plt.figure(figsize=(12,9))
rows, cols = 3, 3

for i in range(1, rows*cols+1):
    img = test_samples[i-1]
    label = pred_labels[i-1]
    fig.add_subplot(rows, cols, i)
    plt.imshow(img.squeeze(), cmap='gray')
    plt.title(f"Predicted: {class_names[label]} \n True: {class_names[test_labels[i-1]]}")
    plt.axis(False)
```

__Plotting Confusion Matrix__ <br>
A confusion matrix is a matrix that summarizes the performance of a machine learning model on a set of test data. It is a means of displaying the number of accurate and inaccurate instances based on the models predictions. 

```python
# getting all predictions in a tensor

y_preds = []
cnn.eval()
with torch.inference_mode(): 
    for X_test, y_test in tqdm(test_dataloader):
        # send data to target device
        X_test, y_test = X_test.to(device), y_test.to(device)
        
        # forward pass
        y_logits = cnn(X_test)
        
        # turn logits to probabilities and then to labels
        y_pred = torch.softmax(y_logits.squeeze(), dim=0).argmax(dim=1)
        
        # put predictions on cpu
        y_preds.append(y_pred.cpu())
        
y_pred_tensor = torch.cat(y_preds)
y_pred_tensor
```
